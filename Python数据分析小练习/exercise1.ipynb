# **第一题**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("人事管理数据.csv",header=0,encoding="GBK")
df.head(5)
```

```python
# 数据透视表
pivot_table = df.pivot_table(index=["性别","部门","学历","职称"],
                             values=['工资'],
                             aggfunc=['mean','max']
                            ).reset_index()
round(pivot_table,0)
```

```python
# 工资水平直方图
sns.set_style('darkgrid')
sns.set(font="Microsoft YaHei")
sns.histplot(df['工资'],bins=10,color='blue',kde=True)
plt.xlabel('工资水平')
plt.ylabel('工资频数');
```

# **第二题**

```python
import seaborn as sns
data = sns.load_dataset('tips')
data.head(5)
```

```python
print(data.shape)
```

```python
sns.set_theme(style='darkgrid')
sns.set(font="Microsoft YaHei")
fig = plt.figure(figsize=(16,16))
ax1 = fig.add_subplot(3,2,1)
ax2 = fig.add_subplot(3,2,2)
ax3 = fig.add_subplot(3,2,3)
ax4 = fig.add_subplot(3,2,4)
ax5 = fig.add_subplot(3,2,5)
ax6 = fig.add_subplot(3,2,6)
sns.scatterplot(data['tip'],data['total_bill'],color='blue',hue=data['smoker'],ax=ax1)
sns.regplot(data['tip'],data['total_bill'],color='green',ax=ax2);
sns.boxplot(data['sex'],data['tip'],color='#8A2BE2',ax=ax3)
sns.boxplot(data['smoker'],data['tip'],color='#8A2BE2',ax=ax4)
sns.violinplot(data['time'],data['tip'],color='#FF7F50',ax=ax5)
sns.violinplot(data['size'],data['tip'],color='#ff7F50',ax=ax6)    
```

# **第三题**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
data = sns.load_dataset('flights')
data.head(5)
```

```python
data['year']
```

```python
data[data['year'] ==1960]
```

```python
sns.set_theme(style='darkgrid')
sns.set(font="Microsoft YaHei")
fig = plt.figure(figsize=(16,12))
fig.subplots_adjust(left=None, bottom=None, right=None, top=None,wspace=None, hspace=0.4)
ax1 = fig.add_subplot(3,1,1)
ax2 = fig.add_subplot(3,1,2)

passengers_year = data.groupby('year')['passengers'].sum()
sns.lineplot(x=data['year'].unique(),y=passengers_year,ax=ax1)
ax1.set_xlabel('年份', fontsize=16)
ax1.set_ylabel('乘客数量', fontsize=16)
ax1.set_title('乘客总数变化趋势', fontsize=20)

passengers_1960 = data[data['year'] ==1960].groupby('month')['passengers'].sum()
sns.histplot(data=data,x=passengers_1960,bins=5,kde=True,ax=ax2)
ax2.set_xlabel('月份', fontsize=16)
ax2.set_ylabel('乘客总数', fontsize=16)
ax2.set_title('1960年乘客人数分布直方图', fontsize=20);
```

```python
fig = plt.figure(figsize=(16,12))
fig.subplots_adjust(left=None, bottom=None, right=None, top=None,wspace=None, hspace=0.4)
ax1 = fig.add_subplot(3,1,1)
ax2 = fig.add_subplot(3,1,2)
ax3 = fig.add_subplot(3,1,3)

passengers_year = data.groupby('year')['passengers'].sum()
sns.lineplot(x=data['year'].unique(),y=passengers_year,ax=ax1)
ax1.set_xlabel('年份', fontsize=16)
ax1.set_ylabel('乘客数量', fontsize=16)
ax1.set_title('乘客总数变化趋势', fontsize=20)

passengers_month = data.groupby('month')['passengers'].sum()#.sort_values(decreasing=False)
sns.barplot(x=data['month'].unique(),y=passengers_month,ax=ax2)
ax2.set_xlabel('月份', fontsize=16)
ax2.set_ylabel('乘客数量', fontsize=16)
ax2.set_title('历年乘客人数变化趋势', fontsize=20);

passengers_1960 = data[data['year'] ==1960].groupby('month')['passengers'].sum()#.sort_values(decreasing=False)
sns.barplot(x=data['month'].unique(),y=passengers_1960,palette='pastel',ax=ax3)
ax3.set_xlabel('月份', fontsize=16)
ax3.set_ylabel('乘客总数', fontsize=16)
ax3.set_title('1960年乘客人数变化趋势', fontsize=20);
```
